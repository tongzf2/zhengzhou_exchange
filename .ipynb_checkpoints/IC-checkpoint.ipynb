```python
# coding: utf-8
import numpy as np
import pandas as pd
import datetime
from tqdm import tqdm
```

```python
from alphalens import performance
from alphalens import plotting
from alphalens import tears
from alphalens import utils
```

```python
#筛掉极值
def mad(factor):
    median = np.median(factor)
    mad = np.median(abs(factor-median))
    high = median + (3 * 1.4826 * mad)
    low = median - (3 * 1.4826 * mad)
    factor = np.where(factor>high, high, factor)
    factor = np.where(factor<low, low, factor)
    return factor
```

```python
#标准化计算
def stand(factor):
    mean = factor.mean()
    std = factor.std()
    return (factor-mean)/std
```

```python
long_factor = []
short_factor = []
```

```python
#计算每日增减量数据
def get_daily_postion_change(df_pos):
    adds = []
    dates = []
    minus = []
    for date, sub in tqdm(df_pos.groupby('日期')):
        sub['增减量.1'] = sub['增减量.1'].astype('float')
        sub['增减量.2'] = sub['增减量.2'].astype('float')
        dates.append(date)
        adds.append(sub['增减量.1'].sum())
        minus.append(sub['增减量.2'].sum())
    result = pd.DataFrame({'date':dates, 'long_delta':adds, 'short_delta':minus})
    result['date'] = pd.to_datetime(result['date'])
#     result.set_index(['date'], inplace=True)
    return result
```

```python
#读取持仓数据
def reshape_data(symbol):
    result = pd.DataFrame()
    for i in range(12):
        n = i+1
        sub = pd.read_csv('month_'+str(n)+'_result_data.csv')
        sub = sub.loc[sub['品种/合约']==symbol]
        result = pd.concat([result, sub], axis=0)
    return result
```

```python
symbol_list = ['甲醇MA', '玻璃FG','短纤PF','纯碱SA','硅铁SF','锰硅SM','PTA','尿素UR','动力煤ZC']
for i in symbol_list:
    df_pos = reshape_data(i)
    df_pos_trading = get_daily_postion_change(df_pos)
    print(df_pos_trading.head())
```

```python
#整合得到每日加减仓量
symbol_list = ['甲醇MA', '玻璃FG','纯碱SA','硅铁SF','锰硅SM','PTA','尿素UR','动力煤ZC']
long_factor = pd.DataFrame()
short_factor = pd.DataFrame()
for i in symbol_list:
    df_pos = reshape_data(i)
    df_pos_trading = get_daily_postion_change(df_pos)
    df_pos_trading['symbol'] = i
    tmp_long = df_pos_trading[['date', 'symbol', 'long_delta']]
    long_factor = pd.concat([long_factor, tmp_long], axis=0)
    tmp_short = df_pos_trading[['date', 'symbol', 'short_delta']]
    short_factor = pd.concat([short_factor, tmp_short], axis=0)
```

```python

multiindex_long_factor = long_factor.set_index(['date','symbol'])
multiindex_long_factor.sort_index(inplace=True)
print(multiindex_long_factor.head())

multiindex_short_factor = short_factor.set_index(['date','symbol'])
multiindex_short_factor.sort_index(inplace=True)
print(multiindex_short_factor.head())
```

```python
multiindex_long_factor['long_delta']=mad(multiindex_long_factor['long_delta'])
multiindex_long_factor['long_delta']=stand(multiindex_long_factor['long_delta'])
singlefactor_series_long = multiindex_long_factor['long_delta'].shift(1)

multiindex_short_factor['short_delta']=mad(multiindex_short_factor['short_delta'])
multiindex_short_factor['short_delta']=stand(multiindex_short_factor['short_delta'])
singlefactor_series_short = multiindex_short_factor['short_delta'].shift(1)
```

```python
# 读取close价格数据
df = pd.read_csv('feature_price.csv', engine='python', parse_dates=['date'], index_col=['date'], infer_datetime_format=True)
df.head()
```

```python
# 计算IC
factor_return_long = utils.get_clean_factor_and_forward_returns(singlefactor_series_long.astype(float), df.astype(float))
print(factor_return_long)
```

```python
factor_return_short = utils.get_clean_factor_and_forward_returns(singlefactor_series_short.astype(float), df.astype(float))
print(factor_return_short)
```
